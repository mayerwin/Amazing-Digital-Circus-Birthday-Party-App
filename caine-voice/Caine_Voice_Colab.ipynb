## 1b) Reference clip has background music? Clean it + audit it
Skip if your clip is already clean. This writes `caine_ref_vocals.wav` (voice only, used for cloning) and `caine_ref_music.wav` (the removed music). Listen to both with the players below to confirm it worked before generating.

```python
!python make_caine_voice.py --isolate
from IPython.display import Audio, display
import os
if os.path.exists('caine_ref_vocals.wav'):
    print('VOICE ONLY (used for Caine):'); display(Audio('caine_ref_vocals.wav'))
if os.path.exists('caine_ref_music.wav'):
    print('REMOVED MUSIC (should be just the backing):'); display(Audio('caine_ref_music.wav'))
```
